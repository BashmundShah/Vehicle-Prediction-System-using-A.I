```python
def replace_space(string):
    return re.sub(' ', '-', string)
```

```python
def download_images(row,directory):
    make = replace_space(row['Make'])
    model = replace_space(row['Model'])
    if row['Version']:
        version = replace_space(row['Version'])
    else:
        return
    date = row['Model Date']
    path = make 
    path = os.path.join(directory, path)
    if(not os.path.isdir(path)):
        os.mkdir(path)
        print(path, " made")
    images = [row['image1'],row['image2'],row['image3'],row['image4'],row['image5']]
    for index,image in enumerate(images):
        imagePath = os.path.join(path, str(row['ID']) + '-' + str(index+1)  +'.jpg')
        try:
            r = requests.get(image, allow_redirects=True)
            open(imagePath, 'wb').write(r.content)
        except Exception as e:
            print(e)
    print(row['ID'], ' images downloaded')
    return
```

```python
ROOT_PATH = 'DATASET'
Training = os.path.join(ROOT_PATH, "TRAINING")
if(not os.path.isdir(Training)):
    os.mkdir(Training)
    print("Training Directory Created")
for i in range(data.shape[0]):
    try:
        download_images(data.iloc[i,:],Training)
    except Exception as e:
        print(e)
```
